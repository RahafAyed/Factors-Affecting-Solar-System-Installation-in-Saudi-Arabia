```python
import pandas as pd
import numpy as np
import seaborn as sns                     
import matplotlib.pyplot as plt 
from plotnine import *
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_excel('datacleaning3V2.xlsx')
df.head(10)
```

```python
df.columns
```

```python
g = ggplot(df, aes(x='Income', y='AMBPS', color='Region')) +labs(
    x="Income",
    y="Avearge Bill in Summer",
    color="Region",
    title="Bills Per Income and Region",
)+ geom_line(stat='summary')
print(g)
```
